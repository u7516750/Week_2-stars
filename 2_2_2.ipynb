
Importing the required libraries




```python
import math
import pandas as pd
import numpy as np
```

From previous question, updating the variable Z

```python
Z_t_1 = 1.108875 # Z_1, Z_2, Z_3, Z_4 at temperature 5000 K
Z_t_2 = 1.456335 # Z_1, Z_2, Z_3, Z_4 at temperature 10000 K
Z_t_3 = 2.271550 #Z_1, Z_2, Z_3, Z_4 at temperature 20000 K
```

```python

K = 8.617333262*(math.pow(10, -5)) # defining the boltzman constant in ev
K
```

```python
s=[5000, 10000, 20000]
g= pd.DataFrame(data = s)
g.columns= ["T"]
g
```

For s = 1, N1/NI

```python
g["N1/NI"]= [1/Z_t_1, 1/Z_t_2, 1/Z_t_3]
g
```

For s= 2, N2/ NI

```python
g["N2/NI"]= g["N1/NI"]*(np.exp(-1/(K*g["T"])))
```

```python
g
```

For s = 3, N3/ NI

```python
g["N3/NI"]= g["N1/NI"]*(np.exp(-2/(K*g["T"])))
g
```

```python
g["N4/NI"]= g["N1/NI"]*(np.exp(-3/(K*g["T"])))
g["N5/NI"]= g["N1/NI"]*(np.exp(-4/(K*g["T"])))
g["N6/NI"]= g["N1/NI"]*(np.exp(-5/(K*g["T"])))
g["N7/NI"]= g["N1/NI"]*(np.exp(-6/(K*g["T"])))
g["N10/NI"]= g["N1/NI"]*(np.exp(-9/(K*g["T"])))
g["N15/NI"]= g["N1/NI"]*(np.exp(-14/(K*g["T"])))

```

```python
g
```

```python
gn=g.transpose()
gn
```
